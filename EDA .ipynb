# Logistic Regression


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
x=df1.Salary.array.reshape(-1,1)
y=df2.Rating.array.reshape(-1,1)
```

```python
y
```

```python
x
```

```python
# Spliting the dataset

from sklearn.model_selection import train_test_split

xtrain,ytrain,xtest,ytest=train_test_split(x,y,test_size=0.20)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(model.predict([[4.0]]))
```

```python

model=LinearRegression()
model.fit(xtrain,xtest)
```

```python
predict=model.predict(ytest)
predict
```

```python
# Overfitting and underfitting situation
model.score(xtest,xtrain)*100
```

```python
model.score(ytest,ytrain)*100
```

```python
# Regression evalution metrices:
#1 MAE
#2 MSE
#3 RMSE
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(ytest,predict))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(ytest,predict))
```

```python
import math
print(math.sqrt(mean_squared_error(ytest,predict)))
```
